```python
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
def get_x(r): return path/'train'/r['fname']
def get_y(r): return r['labels'].split(' ')
```

```python
path = Path()
learn_inf = load_learner(path/'nn_desafio5.pkl', cpu=True)
# Create widgets
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
btn_run = widgets.Button(description='Classify')
```

```python
def predict(pred, pred_idx, probs, lbl_pred):
  k = 0
  idx = []
  for i in pred_idx:
    if i == True:
      idx.append(k)
    k += 1
  aux = ''
  for i in range(len(idx)):
    if len(idx) == 1:
      aux += f'{i+1} - Prediction: {pred[i]}; Probability: {probs[idx[i]]*100:2.2f}'
    elif i == len(idx)-1:
      aux += f'{i+1} - Prediction: {pred[i]}; Probability: {probs[idx[i]]*100:2.2f}'
    else:
      aux += f'{i+1} - Prediction: {pred[i]}; Probability: {probs[idx[i]]*100:2.2f}; '

  lbl_pred.value = aux
```

```python
# Define function for 'Classify' button
def on_click_classify(change):
    lbl_pred.value = ''
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl:
      display(img.to_thumb(400, 400))
    pred,pred_idx,probs = learn_inf.predict(img)
    predict(pred, pred_idx, probs, lbl_pred)
btn_run.on_click(on_click_classify)
```

```python
# Putting back btn_upload to a widget for next cell
btn_upload = widgets.FileUpload()
```

```python
VBox([widgets.Label('Select your image!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
